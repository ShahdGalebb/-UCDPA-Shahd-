```python
from scipy import stats
import pandas as pd
import numpy as np
import pandas_datareader as web
import datetime
import matplotlib.pyplot as plt
#matplotlib inline
```

```python
start = datetime.datetime(1998,1,1)
end = datetime.datetime(2022,1,30)
```

```python
tesla = web.DataReader ("TSLA",'yahoo', start,end)
```

```python
tesla.head()
```

```python
tesla['Open'].plot(label = 'tesla open price', figsize=(15,5))
tesla['Close'].plot(label= 'tesla close price')
tesla['High'].plot(label= 'tesla High price')
tesla['Low'].plot(label= 'tesla Low Price')
plt.legend()
plt.title('Tesla Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
tesla['Volume'].plot(figsize=(15,5))
plt.title('Volume Traded By Tesla')
```

```python
tesla = web.DataReader("TSLA", 'yahoo',start,end)
toyota= web.DataReader ("TYO",'yahoo', start,end)
```

```python
tesla['Open'].plot(label='Tesla',figsize=(15,5))
toyota['Open'].plot(label='Toyota')
plt.ylabel('Stock Price')
plt.title('Stock Prices of Tesla, Toyota')
plt.legend()
```

```python
tesla['Volume'].plot(label= 'Tesla',figsize=(15,5))
toyota['Volume'].plot(label= 'Toyota')
plt.ylabel('Volume Traded')
plt.legend()
```

```python
tesla['Total Traded'] = tesla['Open'] * tesla['Volume']
toyota['Total Traded'] = toyota['Open'] * toyota['Volume']
```
